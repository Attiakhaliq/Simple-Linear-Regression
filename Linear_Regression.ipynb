```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("/content/drive/MyDrive/Salary_dataset.xlsx")
df
```

```python
df=df[["YearsExperience","Salary", ]]
df
```

```python
plt.xlabel("YearsExperience", color="Blue")
plt.ylabel("Salary", color="red")
plt.scatter(df.YearsExperience,df.Salary)
plt.plot(df.YearsExperience,df.Salary)
plt.show()
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
lr.fit(df[["YearsExperience"]],df.Salary)
```

```python
lr.predict([[7]])
lr.coef_
lr.intercept_
```

```python
lr.score(df[["YearsExperience"]],df.Salary)
```

```python
plt.xlabel("YearsExperience", color="Blue")
plt.ylabel("Salary", color="red")
plt.scatter(df.YearsExperience,df.Salary)
plt.plot(df.YearsExperience,lr.predict(df[["YearsExperience"]]))
plt.show()
```
